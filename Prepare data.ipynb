```python
test_dataframe = pd.read_csv("Test.csv")
test_dataframe.head()
```

```python
lab_1 = test_dataframe['LIT101_label'].to_numpy()
lab_2 = test_dataframe['DPIT301_label'].to_numpy()
lab_3 = test_dataframe['LIT301_label'].to_numpy()
sum_lab = np.empty_like(lab_1)
```

```python
for i in range(len(sum_lab)):
    if lab_1[i]=='Anomal' or lab_2[i] == 'Anomal' or lab_3[i]=='Anomal':
        sum_lab[i] = 'Anomal'
    else:
        sum_lab[i] = 'Normal'
```

```python
train_dataframe = pd.read_csv("Train multimode.csv")
train_dateindex = pd.date_range("2015-12-25 10:00:00", 
                                periods=train_dataframe.shape[0], freq="s")
train_dataframe.index = train_dateindex




test_dataframe = pd.read_csv("Test.csv")
test_dateindex = pd.date_range("2015-12-28 10:00:00",
                               periods=test_dataframe.shape[0], freq="s")
test_dataframe.index = test_dateindex

```

```python
train_dataframe
```

```python
new_train = train_dataframe['LIT301'].to_numpy()
new_test = test_dataframe['LIT301'].to_numpy()
```

```python
k=0
for i in train_dataframe.index:
    if i >= pd.to_datetime("26.12.2015 00:00:00") and i <= pd.to_datetime("27.12.2015 10:00:00"):
        new_train[k] += 1000
    k+=1
```

```python
k=0
for i in test_dataframe.index:
    if i >= pd.to_datetime("28.12.2015 11:48:15") and i <= pd.to_datetime("28.12.2015 13:35:12"):
        new_test[k] += 500
    if i >= pd.to_datetime("29.12.2015 06:00:00"):
        new_test[k] += 1000
    if i >= pd.to_datetime("29.12.2015 11:37:04") and i <= pd.to_datetime("29.12.2015 13:21:44"):
        new_test[k] -= 400
    k+=1
```

```python
new_labels = np.zeros(len(test_dataframe), dtype = object)
k = 0
for i in test_dataframe.index:
    if i >= pd.to_datetime("28.12.2015 11:48:15") and i <= pd.to_datetime("28.12.2015 13:35:12"):
        new_labels[k] = 'Anomal'
    elif i >= pd.to_datetime("29.12.2015 11:37:04") and i <= pd.to_datetime("29.12.2015 13:21:44"):
        new_labels[k] = 'Anomal'
    else:
        new_labels[k] = 'Normal'
    k+=1
```

```python
new_tain_dataframe = pd.DataFrame(data = new_train,
                                 columns = ['LIT301 multimode'],
                                 index = train_dataframe.index)
```

```python
new_tain_dataframe
```

```python
new_test_dataframe = pd.DataFrame(data = np.column_stack((new_test, new_labels)),
                                 columns = ['LIT301 multimode', 'LIT301_label'],
                                 index = test_dataframe.index)
```

```python
new_test_dataframe
```

```python
new_tain_dataframe.to_csv('Train multimode.csv', sep = ',')
new_test_dataframe.to_csv('Test multimode.csv', sep = ',')
```
